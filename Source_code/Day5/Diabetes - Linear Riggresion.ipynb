```python
import pandas as pd
input_data = pd.read_csv("diabetes.tab.txt", sep='\t', header=[0])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
diabetes_data = input_data.drop(['Y'], axis=1)

diabates_target = input_data['Y']

X_train, X_test, y_train, y_test = train_test_split(diabetes_data, diabates_target, test_size=0.3, random_state=0)
print(X_train.shape)
print(X_test.shape)

model_diabates = LinearRegression().fit(X_train, y_train)
```

```python
model_diabates.coef_
```

```python
diabetes_data.columns
```

```python
import matplotlib.pylab as plt
%matplotlib inline
import seaborn as sns
cols = ["Y", "BMI", "AGE", "BP"]
sns.pairplot(input_data[cols])
plt.show()
```

```python
predictions = model_diabates.predict(X_train)

plt.scatter(y_train, predictions)

plt.show()
```

```python
from sklearn.metrics import explained_variance_score, mean_squared_error, mean_absolute_error, r2_score

print('explained_variance_score: {}'.format(explained_variance_score(y_train, predictions)))
print('mean_squared_error: {}'.format(mean_squared_error(y_train, predictions)))
print('r2_score: {}'.format(r2_score(y_train, predictions)))
```

```python
predictions = model_diabates.predict(X_test)

plt.scatter(y_test, predictions)

plt.show()
```

```python
from sklearn.metrics import explained_variance_score, mean_squared_error, mean_absolute_error, r2_score

print('explained_variance_score: {}'.format(explained_variance_score(y_test, predictions)))
print('mean_squared_error: {}'.format(mean_squared_error(y_test, predictions)))
print('r2_score: {}'.format(r2_score(y_test, predictions)))
```
